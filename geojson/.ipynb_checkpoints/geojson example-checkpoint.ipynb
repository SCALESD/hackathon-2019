# Plotting geojson data with geopandas
goepandas provides a number of built in data sets that you can leverage.  Let's load in a map of the world with some associated statistics to explore.  We print the first few records in the data set and find that it contains countries with their associated continent, population, and polygon representing their shape.

```python
import geopandas

world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
world.head()
```

Next we'll plot the shapes in the dataset.

```python
world.plot()
```

Now let's look at San Diego.  The following data sets were downloaded from the San Diego Open Data Portal at https://data.sandiego.gov.  We'll use geopandas to read the file in and then plot it.

```python
sandiego = geopandas.read_file('data/san_diego_boundary_datasd.geojson')
sandiego.plot(figsize=(12,12))
```

Let's try combining multiple datasets by overlaying San Diego roads on the map.  First we'll grab the roads geojson from the Open Data Portal, load it, and plot it.

```python
roads = geopandas.read_file('data/roads_datasd.geojson')
roads.plot()
```

With both datasets loaded, let's create a base plot and then overlay the roads on top of it.

```python
base = sandiego.plot(color='white', edgecolor='black', figsize=(12,12))
roads.plot(ax=base)
```

We can take things a step further with Choropleth maps.  These are maps where regions are shaded to represent a proportion or measurement.  Let's do this with the world dataset we loaded at the beginning. 

```python
world = world[(world.pop_est>0) & (world.name!="Antarctica")]
world['gdp_per_cap'] = world.gdp_md_est / world.pop_est
world.plot(column='gdp_per_cap', figsize=(12,12));
```
